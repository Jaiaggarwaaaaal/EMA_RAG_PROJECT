```python
import torch
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import PromptTemplate
# 
# Model names (make sure you have access on HF)

LLAMA2_7B_CHAT = "meta-llama/Llama-2-7b-chat-hf"


selected_model = LLAMA2_7B_CHAT

SYSTEM_PROMPT = """
- You are an AI assistant that answers questions in a friendly manner, based on the given source documents. Here are some rules you always follow:
- Generate human readable output, avoid creating output with gibberish text.
- Generate only the requested output, don't include any other language before or after the requested output.
- Never generate offensive or foul language.
"""

query_wrapper_prompt = PromptTemplate(
    "[INST]<>\n" + SYSTEM_PROMPT + "<>\n\n{query_str}[/INST] "
)
```

```python

!pip install transformers huggingface_hub

# Step 2: Import the required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
# 
# Step 3: Initialize the model without quantization
model_name = "meta-llama/Llama-2-7b-chat-hf"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32,  # Use full precision since we are on CPU
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=1048,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name=selected_model,
    model_name=selected_model,
    device_map="auto",
    # change these settings below depending on your GPU
    model_kwargs={"torch_dtype": torch.float16, "load_in_4bit": True},
)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from llama_index.embeddings.langchain import LangchainEmbedding

lc_embed_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2"
)
embed_model = LangchainEmbedding(lc_embed_model)
```

```python
%pip install llama-index
```

```python
from llama_index.core import ServiceContext
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size = 1024,
    llm = llm,
    embed_model = embed_model

)
```

```python
service_context
```

```python
index = VectorStoreIndex.from_documents(documents, service_context = service_context)
```

```python
query_engine = index.as_query_engine()
```
